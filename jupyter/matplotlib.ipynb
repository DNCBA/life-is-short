```python
# 画折线形图
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,4,5,6]
fig1 =plt.figure(figsize=[5,5])
plt.plot(x,y)
plt.title('titile')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.show()
```

```python
# 画散点图
fig2= plt.figure(figsize=[5,5])
plt.scatter(x,y)
plt.title('titile')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.show()
```

```python
# 画柱状图
fig3= plt.figure(figsize=[5,5])
plt.bar(x,y)
plt.title('titile')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.show()
```

```python
# 画折线形对比图
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y1= [2,3,4,5,6]
y2 = [5,6,7,3,9]
fig4 =plt.figure(figsize=[5,5])
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('titile')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.show()
```

```python
# 多个子图
fig5= plt.figure(figsize=[15,15])
plt.subplot(3,  2,  1) 
plt.scatter(x,y)
plt.title('1-1')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.subplot(3,  2,  2) 
plt.scatter(x,y)
plt.title('1-2')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.subplot(3,  2,  3) 
plt.plot(x,y)
plt.title('2-1')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.subplot(3,  2,  4) 
plt.plot(x,y)
plt.title('2-2')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.subplot(3,  2,  5) 
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('3-1')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.subplot(3,  2,  6) 
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('3-2')
plt.xlabel('x shep')
plt.ylabel('y shep')
plt.show()
```
